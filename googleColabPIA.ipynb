# Uso de GroupBy en un DataFrame: agrupar los datos por diferentes características, como la sucursal, para realizar un análisis como las ventas totales o el producto más vendido por sucursal.

```python
sucursal = df.groupby('Sucursal')
sucursal.first()
```

```python
ventas_suc = df.groupby('Sucursal')['TotalVenta'].sum()
print(f"Las ventas por sucursal son:\n{ventas_suc}")
```

```python
prod_masVendido = df.groupby('Producto')['Cantidad'].sum()
print(f"El producto más vendido es:\n{prod_masVendido.sort_values(ascending=False)}")
```

# Estadísticos Descriptivos: se calcularán los estadísticos para las ventas, como la media, mediana, varianza, etcétera.

```python
ventas_media = df['TotalVenta'].mean()
print(f"La media del total de ventas es:\n{ventas_media}")
```

```python
ventas_mediana = df['TotalVenta'].median()
print(f"La mediana del total de ventas es:\n{ventas_mediana}")
```

```python
ventas_moda = df['TotalVenta'].mode()
print(f"La moda del total de ventas es:\n{ventas_moda}")
```

```python
ventas_desviacion = df['TotalVenta'].std()
print(f"La desviacion estándar del total de ventas es:\n{ventas_desviacion}")
```

```python
ventas_var = df['TotalVenta'].var()
print(f"La varianza del total de ventas es:\n{ventas_var}")
```

```python
ventas_cuartiles = df['TotalVenta'].quantile()
print(f"Los cuartiles del total de ventas es:\n{ventas_cuartiles}")
```

# Gráficas

```python
# importar las librerias
from pandas.core.frame import DataFrame
import matplotlib.pyplot as plt
```

```python
df.groupby('Sucursal')['TotalVenta'].sum().plot(kind='bar')
plt.title('Ventas Totales por sucursal')
plt.xlabel('Sucursal')
plt.ylabel('TotalVenta')
plt.show()
```

```python
df.groupby('Producto')['Cantidad'].sum().plot(kind='line')
plt.title('Cantidad de producto por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.groupby('Producto')['Cantidad'].mean().plot(kind='barh')
plt.title('Promedio de cantidad por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.groupby('Producto')['Cantidad'].var().plot(kind='area')
plt.title('Varianza de cantidad por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.groupby('Producto')['Cantidad'].std().plot(kind='pie')
plt.title('Desviación Estándar de cantidad por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```
